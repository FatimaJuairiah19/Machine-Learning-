```python
#importing library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('carprices.csv')
```

```python
df
```

```python
plt.scatter(df['Car Model'],df['Mileage'],df['Sell Price'],df['Age'])
```

```python
dummies= pd.get_dummies(df['Car Model'])
```

```python
df
```

```python
dummies
```

```python
merged= pd.concat([df,dummies],axis='columns')
```

```python
merged
```

```python
final=merged.drop(['Car Model'],axis='columns')
```

```python
final
```

```python
y=df[('Sell Price')]
```

```python
y
```

```python
x=final.drop(['Sell Price'],axis=1)
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[46000,4,0,1,0,0]])
```
